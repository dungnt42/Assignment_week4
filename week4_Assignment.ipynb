```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X = data.drop(columns=['Species', 'Id'], axis=1).values
y = data['Species'].values
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state = 101)
```

```python
rfc = RandomForestClassifier(n_estimators=100)
dtc = DecisionTreeClassifier()
```

```python
rfc.fit(X_train, y_train)
dtc.fit(X_train, y_train)
```

```python
rfc_predictions = rfc.predict(X_test)
accuracy_score(y_test, rfc_predictions)
print(confusion_matrix(y_test, rfc_predictions))
sns.heatmap(confusion_matrix(y_test, rfc_predictions), annot=True, cmap='YlGnBu')
print(classification_report(y_test, rfc_predictions))
```

```python
dtc_predictions = dtc.predict(X_test)
accuracy_score(y_test, dtc_predictions)
print(confusion_matrix(y_test, dtc_predictions))
print(classification_report(y_test, dtc_predictions))
```

```python
n = [50,100,200,300,400,500]
a = []
for i in n:
    rfc = RandomForestClassifier(n_estimators=i)
    rfc.fit(X_train, y_train)
    rfc_prediction = rfc.predict(X_test)
    rfc_accuracy = accuracy_score(y_test, rfc_prediction)
    a.append(rfc_accuracy)
plt.scatter(n,a)
print(a)
```

# DÃ¹ng GridsearchCv

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "n_estimators": [50, 100, 200, 500],
    "max_depth": [40, 50, 100]
}
model = RandomForestClassifier()
gridcv = GridSearchCV(model, param_grid=param_grid, n_jobs=4, verbose=2)
gridcv.fit(X, y)
```

```python
ele_grid = {
    "n_estimators": [50,100,200,500],
    "max_depth": [30,50,100]
}
rfc_grid = RandomForestClassifier()
rfc_grid_model = GridSearchCV(rfc_grid, param_grid=ele_grid, n_jobs=4, verbose=2)
rfc_grid_model.fit(X,y)
```

```python
rfc_grid_model.best_score_
```

```python
rfc_grid_model.best_params_
```

```python
df = pd.DataFrame(rfc_grid_model.cv_results_)
df.head(5)
```
